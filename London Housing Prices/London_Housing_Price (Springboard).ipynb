**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Create the ratio
def price_ratio(e):
    y1998 = float(e['Average_Price'][e['Year']==1998])
    y2018 = float(e['Average_Price'][e['Year']==2018])
    ratio = [y2018/y1998]
    return ratio
    
    
```

```python
# Test your function
price_ratio(df_grp[df_grp['London_Borough']=='Barking & Dagenham'])
```

```python
# To get ratio for all London_Borough,store the ratios in variable 'all_val' by open an dictionary
all_val = {}
```

```python
# Itearte over a loop who iterate over all the values 
for k in df_grp['London_Borough'].unique():
    borough = df_grp[df_grp['London_Borough'] == k]  #subset the df_grp on 'London_Borough' == k
    all_val[k] = price_ratio(borough)   #make a new entry in all_ratio dictionary 

print(all_val) #call the function 
```

```python
# Above DataFrame we have all ratios of all 'London borough'from 1998-2018, now
# Make a new variable called df_ratios and assign it the result by calling the DataFrame method.

df_ratios = pd.DataFrame(all_val)

```

```python
# To check the new DataFrame, call the .head() method
df_ratios.head()
```

```python
# We will transpose it giving new variable df_ratios_trans, reset the indexes, and check it using .head() method.

df_ratios_trans = df_ratios.T
df_ratios = df_ratios_trans.reset_index()
df_ratios.head()

```

```python
# Rename the 'index' column as 'London_Borough', and '0' column to'2018'.

df_ratios.rename(columns = {'index':'London_Borough', 0:'2018'},inplace = True)
df_ratios.head()
```

```python
# Sort the data in descending order and select the top_15 Boroughs.
# make a new variable above_15 and store data in above_15, and check the new DataFrame.

above_15 = df_ratios.sort_values(by = '2018', ascending = False).head(15)
print(above_15)
```

```python
# Plot the Boroughs that have seen greatest changes in price 
# Make a variable called axi assign the result to filtering above_15 on 'Borough' and '2018', then call plot()
# plot bar graph

axi = above_15[['London_Borough','2018']].plot(kind = 'bar')

axi.set_xticklabels(above_15.London_Borough)
plt.xlabel('London_Borough')
plt.ylabel('Price Increase Factor')
plt.title('Average Housing Prices Between 1998 to 2018')
plt.show()
```
